```python
movies = movies.merge(credits, on = 'title')
```

```python
#genres
#id
#keywords
#title
#overview
#production_companies
#cast
#crew

movies=movies[['id','genres','keywords', 'title', 'overview','cast','crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[1].genres
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def convert3(obj):
    L=[]
    count= 0 
    
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count=+1
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
def director(obj):
    L=[]
    for i in ast.literal_eval(obj):
      if  i['job']== 'Director':
          L.append(i['name'])
        
    return L
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies['crew']
```

```python
movies['overview']
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x ])
movies['overview'] = movies['overview'].apply(lambda x : [i.replace(" ","") for i in x ])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x ])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x ])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x ])
```

```python
movies.head(1)
```

```python
movies['tags']= movies['overview']+movies['cast']+movies['crew']+movies['keywords']+movies['genres']
```

```python
movies.head(1)
```

```python
new_df=movies[['title', 'id' , 'tags']]
```

```python
new_df.head(1)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : " " . join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000, stop_words = 'english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vectors)
```

```python
sim[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = sim[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key = lambda x: x[1]) [1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
